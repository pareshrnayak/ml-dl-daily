**Standard Deviation & Normal Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = [2, 4, 6, 8, 10]
print(data)
```

```python
# Mean Function
def mean(data):
    total = 0
    for x in data:
        total += x
    return total / len(data)
```

```python
# Standard Deviation Function
def standard_deviation(data):
    m = mean(data)
    total = 0

    for x in data:
        total += (x - m) ** 2

    variance = total / len(data)
    return variance ** 0.5  # square root
```

```python
print("Data:", data)
print("Mean:", mean(data))
print("Standard Deviation:", standard_deviation(data))
```

```python
print("NumPy Standard Deviation:", np.std(data))
```

```python
# Generate random data with normal distribution
mu = 50       # mean
sigma = 10    # standard deviation
data_normal = np.random.normal(mu, sigma, 1000)
```

```python
plt.figure()
plt.hist(data_normal, bins=30, density=True, edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
mean_value = np.mean(data_normal)
std_value = np.std(data_normal)

plt.figure()
plt.hist(data_normal, bins=30, density=True, edgecolor='black')

# Mean line
plt.axvline(mean_value, color='red', linewidth=2, label='Mean')

# Standard deviation lines
plt.axvline(mean_value + std_value, color='green', linestyle='--', linewidth=2, label='+1 Std Dev')
plt.axvline(mean_value - std_value, color='green', linestyle='--', linewidth=2, label='-1 Std Dev')

plt.title("Normal Distribution with Mean & Std Dev")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```

# Standard Deviation & Normal Distribution (Intuitive Explanation)

In Machine Learning, understanding **how data is spread** is just as important as
knowing its average.  
Standard Deviation and Normal Distribution help us understand **data variability**
and prepare data correctly before training models.

---

## 1 Standard Deviation (Data Spread)

**What is Standard Deviation?**  
Standard Deviation measures **how far data points are from the mean**.

In simple words:
> It tells us whether values are tightly packed or widely spread.

- Small standard deviation  data points are close to the mean  
- Large standard deviation  data points are far from the mean  

**Formula:**
$$
\sigma = \sqrt{\frac{1}{n} \sum (x - \mu)^2}
$$

Where:
- $x$ = data value  
- $\mu$ = mean  
- $n$ = number of values  

**Why Standard Deviation is important in ML:**  
- Used in **feature scaling (Z-score normalization)**  
- Helps identify **outliers**  
- Controls **model stability**  
- Plays a key role in **overfitting and underfitting**  
- Used internally in many ML and DL algorithms  

---

## 2 Normal Distribution (Bell Curve)

When data follows a normal distribution:
- Most values lie close to the **mean**
- Fewer values lie as we move away from the mean
- The overall shape looks like a **bell**

This pattern appears naturally in many real-world datasets.

**Empirical Rule:**
- ~68% of data lies within **Mean  1 Std Dev**
- ~95% within **Mean  2 Std Dev**
- ~99.7% within **Mean  3 Std Dev**

---

## 3 Understanding the Visualization

**Histogram Bars:**  
- Show how often values occur  
- Taller bars near the center indicate many values close to the mean  

**Mean Line:**  
- Marks the **central location** of the data  

**Standard Deviation Lines:**  
- Show the region where most data points fall  
- Values far outside these lines are **rare and may be outliers**

This visualization tells us:
> The data is centered around the mean, and its spread is controlled by the standard deviation.

---

## 4 Normalization using Mean & Std Dev

In Machine Learning, data is often normalized using:
$$
z = \frac{x - \mu}{\sigma}
$$

This process:
- Centers data around **0**
- Scales data to have **unit variance**
- Makes model training **faster and more stable**
- Helps algorithms converge efficiently

---

##  Key Takeaway

- **Standard Deviation** explains data spread  
- **Normal Distribution** shows how data naturally clusters  
- **Visualization** helps detect outliers and data imbalance  
- **Normalization** prepares data for ML/DL models  

These concepts are essential before applying any Machine Learning or Deep Learning algorithm.
