**Simple Linear Regression (From Scratch)**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 7, 9, 11])
```

```python
w = 0.0   # weight (slope)
b = 0.0   # bias (intercept)

learning_rate = 0.01
epochs = 100
```

```python
def predict(x, w, b):
    return w * x + b
```

```python
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)
```

```python
losses = []

for epoch in range(epochs):

    # Predictions
    y_pred = predict(x, w, b)

    # Loss
    loss = mse_loss(y, y_pred)
    losses.append(loss)

    # Gradients
    dw = (-2 / len(x)) * np.sum(x * (y - y_pred))
    db = (-2 / len(x)) * np.sum(y - y_pred)

    # Update parameters
    w = w - learning_rate * dw
    b = b - learning_rate * db

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: w={w:.4f}, b={b:.4f}, loss={loss:.4f}")
```

```python
print("Final weight (w):", w)
print("Final bias (b):", b)
```

```python
plt.figure()
plt.scatter(x, y, label="Actual Data")
plt.plot(x, predict(x, w, b), label="Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Regression (From Scratch)")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(losses)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Reduction During Training")
plt.show()
```

# Simple Linear Regression (From Scratch)

Simple Linear Regression is one of the **most fundamental algorithms in Machine Learning**.
It is used to model the relationship between **one input feature** and **one output value**
by fitting a straight line.

This experiment implements Linear Regression **from scratch** using only NumPy,
to understand how ML works internally.

---

## 1 What is Simple Linear Regression?

Simple Linear Regression tries to learn a relationship of the form:

$$
y = wx + b
$$

Where:
- $x$ = input feature  
- $y$ = output (target value)  
- $w$ = weight (slope of the line)  
- $b$ = bias (intercept)  

The goal is to find the **best values of $w$ and $b$** so that predictions are close to actual data.

---

## 2 Why Linear Regression is Needed

Linear Regression is used when:
- We want to **predict a numerical value**
- There is a **linear relationship** between input and output

Examples:
- Predicting house prices
- Estimating salary based on experience
- Forecasting sales

It also forms the **foundation for many advanced ML models**.

---

## 3 Model Prediction

For every input value $x$, the model predicts:

$$
\hat{y} = wx + b
$$

Initially, the model starts with **random or zero values** for $w$ and $b$,
so predictions are poor.

---

## 4 Measuring Error (Loss Function)

To measure how wrong the model is, we use **Mean Squared Error (MSE)**.

$$
L = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Why MSE is used:
- Penalizes large errors more
- Always positive
- Smooth and easy to optimize

---

## 5 How the Model Learns (Gradient Descent)

The model learns by **minimizing the loss** using Gradient Descent.

Update rules:

$$
w = w - \alpha \frac{\partial L}{\partial w}
$$

$$
b = b - \alpha \frac{\partial L}{\partial b}
$$

Where:
- $\alpha$ = learning rate  
- Gradients show the **direction of error**

This process repeats over many epochs.

---

## 6 Understanding the Visualization

### Regression Line Plot:
- Dots represent **actual data**
- Line represents **model predictions**
- Better fit  line passes closer to points

### Loss Curve:
- X-axis  epochs
- Y-axis  loss value
- Downward curve  model is learning

---

## 7 Why Training Works

At each step:
1. Model makes predictions  
2. Loss is calculated  
3. Gradients are computed  
4. Parameters are updated  
5. Loss decreases  

Over time, the model finds the **best fitting line**.

---

## 8 Importance in Machine Learning

Simple Linear Regression:
- Builds intuition for ML training
- Explains how weights & bias work
- Forms the base for:
  - Multiple Linear Regression
  - Logistic Regression
  - Neural Networks

Understanding this algorithm is **mandatory before moving to complex models**.

---

##  Key Takeaway

- Linear Regression predicts numerical values
- Uses a straight-line equation
- Learns by minimizing loss
- Gradient Descent updates parameters
- Core foundation of Machine Learning
